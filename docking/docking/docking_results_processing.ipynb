## Docking results analysis

```python
docking_df = pd.DataFrame(columns=['name', 'mode', 'affinity_kcal_mol', 'dist_lb', 'dist_ub'])
path = "./"
with open(path + "all_results.txt", "r") as f:
    for line in f:
        data_dict = {}
        if line[0:6] == 'Ligand':
            try:
                found = re.search('ligands\/(.*)\.pdbqt', line).group(1)
            except AttributeError:
                # not found in the original string
                found = 'no_name'
        if line.count('+') == 3:
            try:
                while 'Vina' not in line:
                    line = next(f)
                    if "Vina" not in line:
                        line_ls = [el for el in line.split(' ') if el != '']
                        data_dict['mode'] = line_ls[0]
                        data_dict['affinity_kcal_mol'] = line_ls[1]
                        data_dict['dist_lb'] = line_ls[2]
                        data_dict['dist_ub'] = line_ls[3]
                        data_dict['name'] = found
                        docking_df = docking_df.append(data_dict, ignore_index=True)
            except:
                continue
```

```python
docking_df['dist_ub'] = docking_df['dist_ub'].str.replace('\n', '')
```

```python
for col in ['mode', 'affinity_kcal_mol', 'dist_lb', 'dist_ub']:
    docking_df[col] = docking_df[col].astype(float)
```

```python
docking_df.to_csv(path + 'all_results_preproc.csv', index=False)
```
